```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel('Covid_Data.xlsx')
df_new = df.groupby("State")["Cured"].sum()
df_new = df_new.sort_values(ascending=False).head(5)
df_new.plot(kind="bar")
plt.xlabel("Active Cases")
plt.ylabel("State")
plt.title("Active Cases Vs State")
plt.show()
```
